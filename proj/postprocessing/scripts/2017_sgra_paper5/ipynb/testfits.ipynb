```python
import ehtplot
from astropy.io import fits
from matplotlib import pyplot as plt
import numpy as np
```

```python
with fits.open('data/img-results/ring1.fits') as hdul:
    ring1 = hdul[0].data
with fits.open('data/img-results/ring2.fits') as hdul:
    ring2 = hdul[0].data
with fits.open('data/img-results/ring3.fits') as hdul:
    ring3 = hdul[0].data
with fits.open('data/img-results/nring.fits') as hdul:
    nring = hdul[0].data
```

```python
print(
    np.max(ring1),
    np.max(ring2),
    np.max(ring3),
    np.max(nring),
)
```

```python
fig, axes = plt.subplots(1,4, figsize=(16,4), sharex=True, sharey=True)
fig.tight_layout()

vmax = 1e-3

axes[0].imshow(ring1, origin='lower', cmap='afmhot_10us', vmin=0, vmax=vmax)
axes[1].imshow(ring2, origin='lower', cmap='afmhot_10us', vmin=0, vmax=vmax)
axes[2].imshow(ring3, origin='lower', cmap='afmhot_10us', vmin=0, vmax=vmax)
axes[3].imshow(nring, origin='lower', cmap='afmhot_10us', vmin=0, vmax=vmax)

for ax in axes:
    ax.set_aspect('equal')

fig.savefig('same_vmax.png', dpi=300)
```
