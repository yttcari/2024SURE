```python
import numpy  as np
import pandas as pd

from matplotlib import pyplot as plt

from common import hallmark as hm
```

```python
def mi(hr, vals, T=1):
    mis = []

    t = hr - hr[0]
    print(t)
    n = int(np.max(t) // T)
    for i in range(n):
        mask = (i * T <= t) & (t < (i+1)*T)
        print(mask)
        m = np.mean(vals[mask])
        s = np.std (vals[mask])
        mis.append(s / m)

    return mis

def calmi(path, hr=3, N=64):
    df = pd.read_csv(path, sep='\t')
    return mi(df.time_hr, df.Ftot)
```
